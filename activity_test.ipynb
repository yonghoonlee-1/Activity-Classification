```python
import glob
import numpy as np

path =r'mlpr20_project_test_data'

filenames_test = sorted(glob.glob(path + "/*.txt"))

input_test = []

for filename in filenames_test: 
    input_test.append(np.loadtxt(filename))
```

```python
sum = len(input_test)
print(sum)
```

```python
print(input_test[0])
```

```python
input_test = np.array(input_test)
```

```python
input_test.shape
```

```python
import pickle
from sklearn.externals import joblib

model = joblib.load('mlpr_term_project.pkl') 
```

```python
prediction_results = []
```

```python
for i in range(sum):
    result = model.predict(np.expand_dims(input_test[i], axis=0))
    result = result[0]
    prediction_results.append(result.argmax()+1)
```
